```python
# simple linear prediction manually

hours = [1,2,3,4,5]
marks = [40,50,65,80,90]

# suppose model: y = 10x + 30
predictions = []

for h in hours:
    y = 10*h + 30
    predictions.append(y)

print(predictions)

# calculate error
error = 0
for i in range(len(marks)):
    error += abs(marks[i] - predictions[i])

print("Total Error:", error)
```
