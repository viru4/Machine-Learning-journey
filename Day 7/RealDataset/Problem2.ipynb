Q1 â€” Relationship check

Scatter plot: Experience vs Salary
Is ML regression possible?
Ans- The relationship is linear with minimal variance, making it ideal for linear regression modeling.

```python
sns.scatterplot(x="YearsExperience", y="Salary", data=df)
plt.show()
```

Q2 â€” Distribution

Plot histogram of salary
Is data skewed?

```python
plt.hist(df["Salary"], bins=10)
plt.title("data distribution of salary")
plt.show()
```

Q3 â€” Outliers
