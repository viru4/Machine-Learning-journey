Q1 â€” Does test preparation help?

Plot: test preparation course vs math score

Which group scores higher?
Ans- the group which has completed the course scores higher
ðŸ‘‰ Use boxplot

```python
sns.boxplot(x="test preparation course", y="math score", data=df)
plt.show()
```

Q2 â€” Who performs better?

Compare: Male vs Female
Ans- feamals are better in reading and writing and men are slightly better in maths

Plot average of math, reading, writing

ðŸ‘‰ Use bar chart

```python
df.groupby("gender")[["math score","reading score","writing score"]].mean().plot(kind="bar")
plt.ylabel("Average Score")
plt.show()
```

Q3 â€” Are subjects related?

Make correlation heatmap

Which subjects are strongly related?
Ans- all subjects are strongely related

```python
corr=df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```
