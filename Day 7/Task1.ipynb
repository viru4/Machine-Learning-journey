Create dataset:

age = [18,19,20,21,22,23,24,25]
salary = [10000,12000,14000,16000,18000,20000,22000,24000]


Make scatter plot â†’ What relationship do you see?

```python
import matplotlib.pyplot as plt
import seaborn as sns
age = [18,19,20,21,22,23,24,25]
salary = [10000,12000,14000,16000,18000,20000,22000,24000]
sns.scatterplot((age, salary))
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
plt.show()
```

form the grapgh we can see as the age is increasing the salary is increasing too
it means +ve relationship
