Q. Does Test Preparation Help?
Ans- Yes Test preparation did help in scoring good marks
students with completed course avg marks=69.69
student who did not completed the course avg marks=64.07

```python
sns.boxplot(x="test preparation course", y="math score", data=df)
print(df.groupby("test preparation course")["math score"].agg(["mean","median","min","max","std"]))
plt.show()
```

Step 5  Gender Comparison
Conclusion:
females score slightly less marks in maths than males
female avg=63.63
male avg= 68.72

```python
sns.boxplot(x="gender", y="math score", data=df)
print(df.groupby("gender")["math score"].mean())
plt.show()
```

Step 6  Lunch Effect
calculate effect of lunch on marks
Conclusion:
people with standard lunch are getting more marks by almost 12 marks

```python
sns.boxplot(x="lunch", y="math score", data=df)
print(df.groupby("lunch")["math score"].agg(["mean","median", "min", "max", "std"]))
plt.show()
```
