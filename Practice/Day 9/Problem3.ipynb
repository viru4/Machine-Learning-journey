🔹 Problem 3 — Feature Scaling (Fitness App)

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler

df = pd.DataFrame({
    "Steps":[2000,4000,10000,15000,30000],
    "Calories":[150,200,350,500,900]
})
```

Tasks

Apply Standardization

Print mean & std after scaling

Try MinMaxScaler also

Compare outputs

👉 Goal: understand effect of scaling

```python
# applying standardization
scaler= StandardScaler()
Steps_scaled= scaler.fit_transform(df)
print("std", Steps_scaled)
print("mean", Steps_scaled.mean())
print("Std", Steps_scaled.std())
```

```python
# MinMax scaling
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
mm_scaled=scaler.fit_transform(df)
print("minMax", mm_scaled)
```
