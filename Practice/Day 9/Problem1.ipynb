Problem 1 — Handling Missing Values (Health Dataset)

You collected gym member data:

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    "Age":[21, 22, np.nan, 25, 60, 23, np.nan],
    "Height":[165, 170, 168, np.nan, 172, 169, 171],
    "Weight":[55, 60, 58, 59, 95, np.nan, 62]
})
```

Tasks

Detect missing values

Decide: mean or median? (think about outliers)

Fill values properly using SimpleImputer

Verify no null values remain

👉 Goal: clean dataset ready for ML

```python
# detecting missing values
print(df.isnull().sum())
imputer=SimpleImputer(strategy="median")
df[["Age", "Weight"]]= imputer.fit_transform(df[["Age", "Weight"]])

imputer=SimpleImputer(strategy="mean")
df[["Height"]]= imputer.fit_transform(df[["Height"]])
print(df)
```

Conclusion:
missing values detected as following
Age       2
Height    1
Weight    1
and mean is selected for Height because it doesnt have any outliers
and median is selected for Age and Weight because they have outliers
and no null values remaining
