🔹 Problem 2 — Categorical Encoding (E-commerce Customers)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({
    "City":["Delhi","Mumbai","Chennai","Delhi","Mumbai","Chennai","Delhi"],
    "Gender":["Male","Female","Female","Male","Male","Female","Male"],
    "Purchased":["Yes","No","Yes","No","Yes","No","Yes"]
})
```

Tasks

Encode target column correctly

Encode input columns correctly

Explain WHY you chose those encodings

Check final dataframe shape

👉 Goal: No text should remain

```python
# Target encoding
le = LabelEncoder()
df["Purchased"] = le.fit_transform(df["Purchased"])

# Feature encoding
df = pd.get_dummies(df, columns=["City","Gender"])

print(df)
print(df.shape) # checking the final dataframe shape
```

conclusion:
no text data remaining in the dataset
