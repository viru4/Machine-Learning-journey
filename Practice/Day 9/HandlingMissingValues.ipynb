```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

df = pd.DataFrame({
    "Age":[22,25,np.nan,29],
    "Salary":[25000,np.nan,32000,40000]
})
print(np.mean(df["Salary"]))
imputer = SimpleImputer(strategy="mean")
df[:] = imputer.fit_transform(df)

print(df)
```
