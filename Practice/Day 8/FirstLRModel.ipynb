```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Create dataset

```python
# dataset
data = {
    "hours":[1,2,3,4,5],
    "marks":[40,50,65,80,90]
}
df = pd.DataFrame(data) #this converts the dataset into table
print(df)
```

3Ô∏è‚É£ Separate input (X) and output (y)

```python
X = df[["hours"]]   # input must be 2D
y = df["marks"]
```

VERY IMPORTANT CONCEPT

ML expects:

ùëã=Features (independent variable)
ùë¶=Target (dependent variable)

Here:
hours ‚Üí feature
marks ‚Üí target
Why double brackets [[ ]] ?
df["hours"]   ‚Üí Series (1D) ‚ùå
df[["hours"]] ‚Üí DataFrame (2D) ‚úÖ
ML models require matrix input:

```python
print(X)
```

4Ô∏è‚É£ Create model

```python
model = LinearRegression()
```

You created an empty brain üß†
Right now it knows nothing:
y=mx+c(m=?,c=?)

5Ô∏è‚É£ Train model (MOST IMPORTANT LINE)

```python
model.fit(X,y)
```

The above line performs:

Training process internally

Predict values

Calculate error (MSE)

Adjust slope & intercept

Repeat many times

Find best line

Mathematically it minimizes:
Loss=1/n‚Äã‚àë(y‚àí(mx+c))2

6Ô∏è‚É£ Predict values

```python
pred = model.predict(X)
```

Now model uses learned equation:
y^=mx+c

7Ô∏è‚É£ Print learned parameters

```python
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```

These are the trained parameters
coef_ ‚Üí slope (how fast marks increase)
intercept_ ‚Üí starting marks
Meaning:
Every extra hour studied ‚Üí marks increase by ~13

8Ô∏è‚É£ Visualization

```python
plt.scatter(X,y,color="blue")
# Plots actual data points

# Blue dots = real students
```

```python
plt.plot(X,pred,color="red")
# Plots learned regression line
# Red line = ML prediction model
```

```python
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.title("Linear Regression")
plt.show()
# Adds labels and displays graph.
```

üß† What the graph means

Blue points = Reality
Red line = Model understanding

If line passes close to points ‚Üí model good
If far ‚Üí model bad
