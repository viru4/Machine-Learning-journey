Problem 7 — Heatmap (Feature Selection)

Create correlation heatmap.

Questions:

1.Which feature affects tip most?

2.Which feature useless?

3.Which feature should be target?

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
corr=df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```

Interpretation
strong relation
total_bill ↔ tip

weak relation
size ↔ tip (moderate)

useless for tip prediction
day / time (categorical low impact)

ao the best feature is total_bill for tip prediction
